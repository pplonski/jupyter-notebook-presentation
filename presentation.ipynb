# Presentation from Jupyter

You can create presentation from Jupyter Notebook.

Mixing code and Markdown is just fantastic:
- computation presentations are great,
- I like to code, I don't like to manually copy-paste images,
- sharing presentation as website or PDF.

```python
from IPython.display import Markdown

greetings = "Hello"
name = "World"
Markdown(f"""

## {greetings} {name}!

You can mix Markdown with variables thanks to `IPython.display.Markdown()`.

""")
```

## Simple plot

```python
from matplotlib import pyplot as plt
import random

points_count = 10
color = "blue"
x = [random.gauss(0, 1) for _ in range(points_count)]
y = [random.gauss(0, 1) for _ in range(points_count)]

plt.figure(figsize=(10, 7))
_ = plt.plot(x, y, '.', markersize=14, color=color)
```
